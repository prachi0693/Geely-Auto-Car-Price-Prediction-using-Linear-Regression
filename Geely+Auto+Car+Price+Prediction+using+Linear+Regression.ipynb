# A. Importing Packages and Understanding Data

```python
# Importing all required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importing dataset
CarPrice = pd.read_csv('CarPrice_Assignment.csv')
```

```python
#Inspecting dataset
CarPrice.head(5)
```

```python
CarPrice.info() # There seems to be no null values in the given dataframe
```

```python
# Converting all the CarName variable values to lower case
CarPrice['CarName'] = CarPrice.CarName.astype(str).str.lower()
```

```python
# Extracting Company Names and Model Names from the column CarName
CarPrice[['Car_Company_Name', 'Car_Model_Name']] = CarPrice.CarName.str.split(' ', n=1, expand = True)
```

```python
CarPrice.head(5) # We can see- two new columns are added in the dataframe
```

```python
# Inspecting the dataframe
CarPrice.describe()
```

# B. Data Cleanup

```python
# Dropping Car_Model_Name column, as it will not be used in further analysis
CarPrice.drop(['Car_Model_Name', 'CarName'], axis =1, inplace = True)

# Also dropping car_ID as it is just a number
CarPrice.drop(['car_ID'], axis =1, inplace = True)
```

```python
# Correcting misspelt Car Company Names
CarPrice.loc[CarPrice.Car_Company_Name == 'maxda', 'Car_Company_Name'] = 'mazda' 
CarPrice.loc[CarPrice.Car_Company_Name == 'alfa-romero', 'Car_Company_Name'] = 'alfa-romeo' 
CarPrice.loc[CarPrice.Car_Company_Name == 'Nissan', 'Car_Company_Name'] = 'nissan' 
CarPrice.loc[CarPrice.Car_Company_Name == 'porcshce', 'Car_Company_Name'] = 'porsche' 
CarPrice.loc[CarPrice.Car_Company_Name == 'toyouta', 'Car_Company_Name'] = 'toyota' 
CarPrice.loc[CarPrice.Car_Company_Name == 'vokswagen' , 'Car_Company_Name'] = 'volkswagen' 
CarPrice.loc[CarPrice.Car_Company_Name == 'vw' , 'Car_Company_Name'] = 'volkswagen' 
```

```python
CarPrice.Car_Company_Name.unique()
# As can be seen all the misspelt company names have been corrected
```

# D. Data Preparation

```python
#Correlations observed from the scatter plot and heat map-
# 1. There is high correlation between wheelbase, carlength, carwidth, curbweight and enginesize
# 2. Also there is high correlation between citympg and highwaympg
# Therefore, dropping carlength, carwidth, curbweight and highwaympg
```

```python
CarPrice.drop(['carlength','carwidth','curbweight','highwaympg'], axis =1, inplace = True)
```

# E. Creating dummy variables for Categorical Variables


```python
#Inspecting unique values in each variable present in the dataset
column_list = CarPrice.columns.values.tolist()
for col in CarPrice:
     print(col,'--' , CarPrice[col].unique())
```

```python
# Converting categorical variables with two levels
CarPrice['fueltype'] = CarPrice['fueltype'].map({'gas': 1, 'diesel': 0})
CarPrice['aspiration'] = CarPrice['aspiration'].map({'std': 1, 'turbo': 0})
CarPrice['doornumber'] = CarPrice['doornumber'].map({'two': 1, 'four': 0})
CarPrice['enginelocation'] = CarPrice['enginelocation'].map({'front': 1, 'rear': 0})
```

```python
# Converting categorical variables with more than two levels
df = pd.get_dummies(CarPrice)
df.columns.sort_values()
```

# F. Rescaling the numerical variables by Normalisation

```python
# identifying the columns to be normalised
cols = ['horsepower', 'peakrpm', 'price', 'stroke', 'symboling', 'wheelbase', 'carheight', 'enginesize'
, 'boreratio', 'citympg', 'compressionratio']
# Normalising only the numeric fields 
ndf = df[cols].apply(lambda x: (x-np.mean(x))/ (max(x) - min(x)))
ndf.head()

df['symboling'] = ndf['symboling']
df['carheight'] = ndf['carheight']
df['enginesize'] = ndf['enginesize']
df['boreratio'] = ndf['boreratio']
df['stroke'] = ndf['stroke']
df['price'] = ndf['price']
df['compressionratio'] = ndf['compressionratio']
df['horsepower'] = ndf['horsepower']
df['peakrpm']= ndf['peakrpm']
df['citympg'] = ndf['citympg']
df['wheelbase'] = ndf['wheelbase']
df
```

# G. Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
df_train.head()
```

```python
#Dividing into X and Y sets for the model building
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
#Building our model
#Using the LinearRegression function from SciKit Learn for its compatibility with RFE
```

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```
